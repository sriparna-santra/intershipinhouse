```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.array([340,1080,640,880,990,510])
y=np.array([500,1700,1100,800,1400,500])

plt.scatter(x,y,color='r',marker="*",label="Actual data")
n=np.size(x)
x_bar= np.mean(x)
y_bar= np.mean(y)

ss_xy= np.sum(x * y) - n * x_bar * y_bar
ss_xx= np.sum(x * x) - n * x_bar * x_bar

b1= ss_xy / ss_xx
b0= y_bar - b1 * x_bar

print(f"b1 is : {b1} and b0 is : {b0}")
y_pred=b1*x+b0

# print(y_pred)
# print(x)
plt.plot(x,y_pred,color="g",label="regression line")

x_new=790
y_new=b1*x_new+b0
plt.text(x_new+30,y_new-20,y_new)

plt.scatter(x_new,y_new,color='b',marker="o",label="Actual data")
print(f"predicted rent for area {x_new} : {y_new : .2f}")

plt.xlabel("Area")
plt.ylabel("Rent")
plt.title("House price prediction")
plt.grid(alpha=0.3)
plt.show()
```
