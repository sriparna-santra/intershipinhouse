```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return (1/(1+np.exp(-x)))
```

```python
def model(x,y,learning_rate,iterations):
    a1=0
    a0=0
    n=np.size(x)
    cost_list=[]
    for i in range(iterations):
        z=a1*x+a0
        a=sigmoid(z)
        cost= -(1/n)*np.sum( y*np.log(a) + (1-y)*np.log(1-a) )
        da1=(1/n)*np.sum((a-y)*x)
        da0=(1/n)*np.sum(a-y)
        a1=a1-learning_rate*da1
        a0=a0-learning_rate*da0
        cost_list.append(cost)
    return a1,a0,cost_list
    
```

```python
x=np.array([22,25,47,52,46,56,55,60,62,61,18,28,27,29,49])
y=np.array([0,0,1,0,1,1,0,1,1,1,0,0,0,0,1])


iterations=10000
learning_rate=0.000001

a1,a0,cost_list=model(x,y,learning_rate,iterations)
print("a0 = ",a0,"   a1 = ",a1)

plt.plot(range(iterations), cost_list)
plt.title("Iterations vs Cost")
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.show()
```

```python
z=a1*x+a0
a = sigmoid(z)
print(z)
print(x,a)

plt.scatter(x,y,color='r',marker='*',label="Actual output")
plt.scatter(x,a,color='g',marker='+',label="predicted output")
plt.xlabel('age')
plt.ylabel('buy')
plt.title('insurance')
plt.legend()

xnew=45
z=a1*xnew+a0
b=sigmoid(z)

plt.scatter(xnew,b,color='b',marker='*',label="unknown input output")
print(b)
plt.text(xnew-0.1,b-0.05,"unknown input output %.2f"%b)
plt.legend()
plt.show()
```

```python

if b>0.5:
    print("buy insurance")
else:
    print("doesnot buy insurance")  
```
